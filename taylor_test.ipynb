```python
import numpy as np
import matplotlib.pyplot as plt

def ziel_funktion(x):
    return np.sin(x) * np.exp(-0.1 * x**2)

x = np.linspace(-5, 5, 100)
y = ziel_funktion(x)

fig, ax = plt.subplots(figsize=(8, 6))


ax.plot(x, y, label='Ziel Funktion', color='blue')
ax.set_title('Plot der Ziel Funktion')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')


plt.show()
```
