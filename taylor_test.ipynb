```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt


x = sp.Symbol("x")
f = sp.sin(x)

f_strich = sp.diff(f, x)
f_2_strich = sp.diff(f, x, 2)

print(f"Die Funktion: {f}")
print(f"Die Ableitung: {f_strich}")
print(f"Die zweite Ableitung: {f_2_strich}")

f_num = sp.lambdify(x, f, "numpy")
f_strich_num = sp.lambdify(x, f_strich, "numpy")

x_werte = np.linspace(-2*np.pi, 2*np.pi, 200)
y_werte = f_num(x_werte)
y_strich_werte = f_strich_num(x_werte)



fig, ax = plt.subplots(figsize=(8, 6))


ax.plot(x_werte, y_werte, label=f"Original Funktion: ${sp.latex(f)}$")
ax.plot(x_werte, y_strich_werte, label=f"Erste Ableitung: ${sp.latex(f_strich)}$")
ax.set_title("Plot der Ziel Funktion")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.legend(loc= "lower right")

plt.show()
```
