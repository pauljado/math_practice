```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

ableitungen = np.arange(1, 5)

x = sp.Symbol("x")
f = sp.sin(x)

x_werte = np.linspace(-2*np.pi, 2*np.pi, 200)


fig, ax = plt.subplots(figsize=(8, 6))

for n in ableitungen:
    f_n_strich = sp.diff(f, x, n)
    f_n_strich_num = sp.lambdify(x, f_n_strich, "numpy")
    y_n_strich_werte = f_n_strich_num(x_werte)
    print(f"Die {n} te Ableitung: {f_n_strich}")
    ax.plot(x_werte, y_n_strich_werte, label=f"{n}. Ableitung: {f_n_strich}")

ax.set_title("Plot der Ziel Funktion")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.axhline(0, ls=":", lw= 1, color="black")
ax.axvline(0, ls=":", lw= 1, color="black")
ax.legend(loc= "lower right")

plt.show()
```
