```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#dx = input()
dx = 0.001

x = np.linspace(0, 10)
y = x**2
x_dx = np.arange(dx, 10 + dx, dx)
y_dx = x_dx**2

# print(x_dx)
#print(y_dx)

riemann_bars = y_dx * dx
riemann_summe = riemann_bars.sum()
#print(riemann_summe)

integral = 10**3 / 3
#print(np.round(integral, 4))

x_1 = np.array([10, 5, 2, 1, 0.5, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001])
y_summe = []
for n in x_1:
    x_dx_1 = np.arange(n, 10 + n, n)
    y_dx_1 = x_dx_1**2
    bars = y_dx_1 * n
    summe = bars.sum()
    y_summe.append(summe)

x_1_reverse = x_1[::-1]

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

sns.lineplot(x=x, y=y, color="orange", ax=ax[0])
ax[0].bar(x_dx, y_dx, width=dx, alpha=0.5, align="center")
ax[0].set_title(f"Riemann Summe (dx={dx}): {np.round(riemann_summe, 4)}")


sns.lineplot(x=x_1_reverse, y=y_summe, ax=ax[1])
ax[1].axhline(integral, color="orange", linestyle="--", label="Tatschliches Integral")
ax[1].set_xscale("log")
ax[1].set_title("Riemann Summe vs. dx")
ax[1].legend()


plt.show()
```
